```python
import pdfplumber

pdf_path = "/content/1435.pdf"

text = ""
with pdfplumber.open(pdf_path) as pdf:
    for page in pdf.pages:
        text += page.extract_text()

# Now, 'text' contains the extracted text from the entire PDF.
```

```python
import re
paragraphs = re.split(r'\n\n', text)
formatted_text = '\n\n'.join(paragraph.strip() for paragraph in paragraphs)
text = formatted_text
text
```

```python
from summarizer import Summarizer,TransformerSummarizer
```

```python
bert_model = Summarizer()
bert_summary = ''.join(bert_model(text, min_length=100))
```

```python
print(bert_summary)
```

```python
GPT2_model = TransformerSummarizer(transformer_type="GPT2",transformer_model_key="gpt2-medium")
GPT2 = ''.join(GPT2_model(text, min_length=100))
```

```python
print(GPT2)
```

```python
model = TransformerSummarizer(transformer_type="XLNet",transformer_model_key="xlnet-base-cased")
XLNet = ''.join(model(text, min_length=100))
```

**For this pdf, our XLNet works the best**

```python
import evaluate
rouge = evaluate.load('rouge')
references = [reference]
generated_bert = [bert_summary]
generated_xlnet = [XLNet]
generated_gpt2 = [GPT2]

rogue = evaluate.load('rouge')
result_bert = rouge.compute(predictions=generated_bert, references=references,use_aggregator=True)
result_XLNet = rouge.compute(predictions=generated_xlnet, references=references,use_aggregator=True)
result_GPT2 = rouge.compute(predictions=generated_gpt2, references=references,use_aggregator=True)

print("Bert:" , result_bert)
print("XLNet:" , result_XLNet)
print("GPT2:" , result_GPT2)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming you have values for result_bert, result_XLNet, and result_GPT2
# Replace these with your actual data

# ROUGE scores for each model
models = ['Bert', 'XLNet', 'GPT2']
rouge_metrics = ['rouge1', 'rouge2', 'rougeL', 'rougeLsum']
scores = {
    'Bert': result_bert,
    'XLNet': result_XLNet,
    'GPT2': result_GPT2
}

# Initialize values for each metric
metric_values = {metric: [scores[model][metric] for model in models] for metric in rouge_metrics}

# Set up the bar positions
bar_width = 0.2
bar_positions = np.arange(len(rouge_metrics))

# Create a grouped bar plot for all metrics
plt.figure(figsize=(8.5, 6.5))

# Specify colors for each model
colors = {'Bert': 'red', 'XLNet': 'blue', 'GPT2': 'green'}

for j, model in enumerate(models):
    plt.bar(bar_positions + j * bar_width, [metric_values[metric][j] for metric in rouge_metrics], bar_width, label=model, color=colors[model])

plt.xlabel('ROUGE Metrics')
plt.ylabel('ROUGE Score')
plt.title('ROUGE Scores for Different Models and Metrics')
plt.xticks(bar_positions + bar_width * (len(models) - 1) / 2, rouge_metrics)
plt.legend()
plt.ylim(0, 1.0)  # Set the y-axis range (adjust as needed)

plt.show()
```

```python
for metric in rouge_metrics:
    plt.plot(models, metric_values[metric], marker='o', label=f'ROUGE-{metric}')

plt.xlabel('Models')
plt.ylabel('ROUGE Score')
plt.title('ROUGE Scores for Different Metrics and Models')
plt.legend()
plt.grid(True)
plt.ylim(0, 1.0)  # Set the y-axis range (adjust as needed)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
# Extract the ROUGE scores into a matrix
score_matrix = [[scores[model][metric] for metric in rouge_metrics] for model in models]

# Create a heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(score_matrix, annot=True, fmt=".2f", cmap="YlGnBu", xticklabels=rouge_metrics, yticklabels=models)
plt.xlabel('ROUGE Metrics')
plt.ylabel('Models')
plt.title('ROUGE Scores Heatmap')
plt.show()
```
